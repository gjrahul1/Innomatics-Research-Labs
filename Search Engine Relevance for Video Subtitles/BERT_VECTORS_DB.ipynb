```python
def connect_db(db_path):
  return sqlite3.connect(db_path)

def fetch_subtitles(conn, sample_percentage = 0.1):
  cursor = conn.cursor()
  cursor.execute("SELECT num, content FROM zipfiles")
  rows = cursor.fetchall()

  if sample_percentage < 1.0:
    rows = random.sample(rows, int(len(rows) * sample_percentage))

  return rows
```

```python
import sqlite3

# Path to your database
DB_PATH = "/content/drive/MyDrive/Files/Datasets/Movie_db/eng_subtitles_database.db"

def get_movie_titles(db_path, limit=50):
    """Fetch movie/show titles from the subtitles database."""
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Query to get unique movie titles from the 'name' column
    cursor.execute("SELECT DISTINCT name FROM zipfiles LIMIT ?", (limit,))
    results = cursor.fetchall()

    conn.close()

    # Extract names
    movie_titles = [row[0] for row in results]
    return movie_titles

# Fetch and display 50 sample movie/show titles
movies = get_movie_titles(DB_PATH)
print("🎬 Sample Movies & Shows in DB:")
for i, title in enumerate(movies, 1):
    print(f"{i}. {title}")
```

```python
def decompress_subtitle(binary_content):
    """
    Decompress subtitle content.
    First attempts to extract text from a ZIP archive.
    If that fails, falls back to zlib decompression.
    If binary_content is a string, it is converted to bytes using 'latin-1'.
    """
    if isinstance(binary_content, str):
        binary_content = binary_content.encode('latin-1')

    # Try ZIP extraction
    try:
        with zipfile.ZipFile(BytesIO(binary_content)) as zf:
            file_names = zf.namelist()
            if file_names:
                with zf.open(file_names[0]) as file:
                    return file.read().decode('latin-1')
    except Exception:
        pass  # ZIP extraction failed; proceed to zlib

    # Try zlib decompression
    try:
        return zlib.decompress(binary_content).decode('latin-1')
    except Exception:
        pass

    return None
```

```python
def preprocess_text(text):
    """
    Remove UTF-8 BOM artifacts, timestamps (e.g., [00:01:23,456]), and extra whitespace.
    """
    text = text.replace("ï»¿", "")    # Remove BOM artifact if present
    text = text.lstrip('\ufeff')       # Remove Unicode BOM if exists
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
def convert_ass_to_srt(text, sub_id):
    """
    If the subtitle text appears to be in ASS/SSA format (detected by markers like "ScriptType:" or "WrapStyle:"),
    attempt to convert it to SRT using pysubs2. If conversion fails with "No suitable formats",
    print a message and return the original text.
    """
    if "ScriptType:" in text or "WrapStyle:" in text:
        try:
            subs = pysubs2.SSAFile.from_string(text)
            if not subs.events or len(subs.events) == 0:
                return text
            srt_text = subs.to_string("srt")
            return srt_text
        except Exception as e:
            error_message = str(e)
            if "No suitable formats" in error_message:
                print(f"Skipping conversion for subtitle {sub_id}: {error_message}")
                return text
            else:
                print(f"Error converting ASS/SSA to SRT for subtitle {sub_id}: {error_message}")
                return text
    return text
```

```python
def chunk_text(text, chunk_size=500, overlap=50):
    """
    Divide long text into smaller chunks.
    Splits based on words, with each chunk having up to 'chunk_size' words and overlapping by 'overlap' words.
    """
    words = text.split()
    if len(words) <= chunk_size:
        return [text]
    chunks = []
    start = 0
    while start < len(words):
        chunk = " ".join(words[start: start + chunk_size])
        chunks.append(chunk)
        start += (chunk_size - overlap)
    return chunks
```

```python
def process_subtitle_row(row, chunk_size=500, overlap=50):
    """
    Process a single subtitle row (num, content).
    Returns a list of tuples (chunk_id, processed_chunk_text).
    """
    sub_id, binary_content = row
    text = decompress_subtitle(binary_content)
    if text is None:
        print(f"Failed to decompress subtitle {sub_id}")
        return []
    text = preprocess_text(text)
    text = convert_ass_to_srt(text, sub_id)
    chunks = chunk_text(text, chunk_size, overlap)
    return [(f"{sub_id}_{idx}", chunk) for idx, chunk in enumerate(chunks)]

def process_all_subtitles(db_path, sample_percentage=0.3, chunk_size=500, overlap=50):
    """
    Connect to the database, fetch subtitles (optionally a sample),
    process each subtitle (decompression, cleaning, conversion, chunking),
    and return a list of tuples (chunk_id, processed_chunk_text).
    """
    conn = connect_db(db_path)
    rows = fetch_subtitles(conn, sample_percentage)
    conn.close()

    processed_all = []
    for row in tqdm(rows, desc="Processing Subtitles"):
        processed_chunks = process_subtitle_row(row, chunk_size, overlap)
        processed_all.extend(processed_chunks)
    return processed_all

# --- Load SentenceTransformer Model on GPU ---
import torch
from sentence_transformers import SentenceTransformer

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🚀 Using device: {device}")

model = SentenceTransformer("all-MiniLM-L6-v2").to(device)


def compute_embeddings_in_batches(model, documents, batch_size=128):
    """
    Compute embeddings in batches using GPU.
    Returns a NumPy array of embeddings.
    """
    embeddings = []
    start_time = time.time()

    for i in tqdm(range(0, len(documents), batch_size), desc="Computing Embeddings"):
        batch_docs = documents[i:i+batch_size]
        with torch.no_grad():
            batch_embeddings = model.encode(batch_docs, convert_to_numpy=True)
        embeddings.extend(batch_embeddings)

        # Estimate remaining time
        elapsed = time.time() - start_time
        batches_done = (i // batch_size) + 1
        avg_time = elapsed / batches_done
        remaining_batches = (len(documents) - i) // batch_size
        time_left = remaining_batches * avg_time
        minutes, seconds = divmod(int(time_left), 60)
        print(f"⏳ Estimated Time Left: {minutes} min {seconds} sec")

    return np.array(embeddings)
```

```python
print("🔄 Extracting and processing subtitles from the database...")
processed_subtitles = process_all_subtitles(DB_PATH, sample_percentage=SAMPLE_PERCENTAGE,
                                            chunk_size=CHUNK_SIZE, overlap=OVERLAP)
print(f"✅ Processed {len(processed_subtitles)} subtitle chunks.")
```

```python
# Separate IDs and documents
ids, documents = zip(*processed_subtitles)
```

```python
print("🔄 Computing embeddings for subtitle chunks on GPU...")
embeddings = compute_embeddings_in_batches(model, list(documents), batch_size=BATCH_SIZE)

if len(embeddings) != len(documents):
    raise ValueError(f"Mismatch: Documents: {len(documents)}, Embeddings: {len(embeddings)}")
```

```python
import whisper
import torch
import chromadb
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 🔄 Load ChromaDB and SentenceTransformer Model
CHROMA_DB_PATH = "/content/drive/MyDrive/subtitle_chromadb"
chroma_client = chromadb.PersistentClient(path=CHROMA_DB_PATH)
collection = chroma_client.get_collection(name="subtitles")

# 🔄 Load SentenceTransformer Model on GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
model = SentenceTransformer("all-MiniLM-L6-v2").to(device)

print("✅ ChromaDB and Model Loaded Successfully!")

def transcribe_audio_whisper(audio_path):
    """
    Uses OpenAI's Whisper model to transcribe an audio file.
    """
    try:
        model = whisper.load_model("base")  # You can switch to "small", "medium", or "large" if needed
        result = model.transcribe(audio_path)
        transcript = result["text"].strip()  # Remove extra spaces

        if not transcript:
            print("❌ Transcription failed: No text generated!")
            return None

        print("\n📝 **Transcription Success:**\n", transcript)
        return transcript
    except Exception as e:
        print(f"❌ Error in transcription: {e}")
        return None

def find_movie_by_subtitle(transcript, top_k=5, similarity_threshold=0.5):
    """
    Searches for the most relevant movies based on the transcribed audio.
    Retrieves matching subtitles, extracts movie/show titles, and suggests movies.
    """
    if not transcript or not transcript.strip():
        print("⚠️ No valid transcript provided! Please check the audio file.")
        return

    print("\n🔍 Searching for related movies...")

    try:
        # Convert query to embedding
        query_embedding = model.encode(transcript.strip(), convert_to_numpy=True).reshape(1, -1)

        # Query ChromaDB
        results = collection.query(
            query_embeddings=[query_embedding.flatten().tolist()],
            n_results=top_k
        )

        if not results or not results.get("ids") or not results["ids"][0]:
            print("❌ No matching subtitles found!")
            return

        # Extract embeddings and compute cosine similarity
        subtitle_embeddings = np.array(results["embeddings"][0])
        similarities = cosine_similarity(query_embedding, subtitle_embeddings)[0]

        # Sort results by similarity
        sorted_indices = np.argsort(similarities)[::-1]

        suggested_movies = set()

        print("\n🎬 Top Related Movies:")
        for rank, index in enumerate(sorted_indices):
            sub_id = results["ids"][0][index]
            subtitle_text = results["metadatas"][0][index]["text"]
            similarity_score = similarities[index]

            # Extract movie title from metadata (if available)
            movie_title = results["metadatas"][0][index].get("movie_title", "Unknown Movie")

            if similarity_score >= similarity_threshold:
                suggested_movies.add(movie_title)  # Add unique movie names
                print(f"\n🎥 **Suggested Movie {rank+1}:** {movie_title} (Similarity: {similarity_score:.2f})")
                print(f"📜 Subtitle: {subtitle_text}")

        if not suggested_movies:
            print("❌ No matching movies found!")

    except Exception as e:
        print(f"❌ Error during movie search: {e}")

# 🚀 **Run Transcription + Movie Search**
AUDIO_PATH = "/content/drive/MyDrive/Files/Datasets/FRIENDS/query_audio.mp3"
transcript_text = transcribe_audio_whisper(AUDIO_PATH)

if transcript_text:
    find_movie_by_subtitle(transcript_text, top_k=10, similarity_threshold=0.5)
else:
    print("❌ No transcript generated. Check the audio file!")
```
